```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ["Population of Vatican City", "My article views"]
y = [825, 1250]
plt.style.use('ggplot')
plt.bar(x, y)
plt.title("People vs Category")
plt.xlabel("Category")
plt.ylabel("Number of people")
plt.savefig('views.png', bbox_inches='tight')
```
